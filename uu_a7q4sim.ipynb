```python
import numpy as np
import matplotlib.pyplot as plt

n = 100000
rho = 0.8

#covariance matrix
Sigma_pos = np.array([[1,  rho],
                      [rho, 1]])

Sigma_neg = np.array([[1, -rho],
                      [-rho, 1]])

#half positive half negative
Z = np.random.binomial(1, 0.5, n)

#generate samples
XY_pos = np.random.multivariate_normal([0,0], Sigma_pos, size=n)
XY_neg = np.random.multivariate_normal([0,0], Sigma_neg, size=n)

XY = XY_pos * (1 - Z[:,None]) + XY_neg * Z[:,None]

X = XY[:,0]
Y = XY[:,1]

print("Sample covariance:", np.cov(X, Y, bias=True)[0,1])

plt.scatter(X, Y, s=2, alpha=0.1)
plt.title("Cov  0 but X,Y clearly dependent")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
